```python
import numpy as np
path_lyrics = 'C:/playdata/DL_project/music_resting_place_ver_dh/3. 노래 가사/final_lyrics/'
ly_df = pd.read_csv(path_lyrics+'translated_lyrics_v2.csv')
ly_df['link'] = np.nan
ly_df['duble_check'] = np.nan
ly_df
```

```python
from selenium import webdriver
from selenium.common.exceptions import WebDriverException as WDE
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from bs4 import BeautifulSoup

## 준비 설정
# 접속 url
origin_url = "https://www.youtube.com"
url = "https://www.youtube.com/results?search_query="

# 검색 키워드
#path = 'C:/playdata/DL_project/music_resting_place_ver_dh/3. 노래 가사/'
for i in range(len(ly_df)) : 
    keyword = (ly_df['Unnamed: 0'].values[i]).replace(' ', '+')

    # 스크롤을 어디까지 내리는지 기준 
    finish_line = 10000

    # 유튜브 홈페이지 접속
    path = 'C:/playdata/DL_project/music_resting_place_ver_dh/0. dataset/chromedriver_win32/chromedriver.exe'
    browser = webdriver.Chrome(path) # 개인에게 맞는 경로 수정 필요
    browser.get(url + keyword)
    time.sleep(1)

    browser.execute_script('window.scrollTo(0, 500)') # 스크롤

    # 정보 가져오기
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, 'html.parser') # lxml

    # 첫 번째 컨텐츠의 제목, 채널명, url 정보 가져오기
    # 첫 번째 컨텐츠 관련 부분 html 구조 떼어내기
    # find: 해당 정보의 첫 번째 요소만 가져오기
    test = soup.find("ytd-video-renderer", attrs={"class":'style-scope ytd-item-section-renderer'})

    # 콘텐츠 모든 정보 가져오기
    title = test.find("yt-formatted-string", attrs={"class":'style-scope ytd-video-renderer'}).get_text()
    content_url = origin_url + test.find("a", attrs={"id":'thumbnail'})["href"]

    # 데이터 프레임에 값 추가
    ly_df['link'][i] = content_url
    ly_df['duble_check'][i] = title 

ly_df
ly_df.to_csv('C:/playdata/DL_project/music_resting_place_ver_dh/3. 노래 가사/final_lyrics/translated_lyrics_and_link_v2.csv')
```
