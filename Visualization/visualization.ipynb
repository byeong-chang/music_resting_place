```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/DL_project/VIsualization/DL.csv')
df
```

```python
# train , test 데이터 과적합 여부를 확인하기 위한 시각화

fig, ax3 = plt.subplots(figsize = (16,9)) # axes3 생성

bar_width = 0.35 # 막대 너비 설정
alpha = 0.5 # 투명도 설정
plt.figure(figsize = (100,100)) # figure 사이즈 설정
N = len(df) # p4 를 그릴 position 생성
index = np.arange(N)
ax3.set_ylabel("train")
#ax3.xaxis.set_tick_params(rotation=45) # ax3의 x 눈금 45도 각도로 출력
p3 = ax3.bar(df['model'],df['before_acc'], bar_width, alpha=alpha, color='b') 
ax3.set_ylim([0.3,0.85]) # y축을 제한하여 표를 더 보기 좋게 만듬

ax4 = ax3.twinx() # axes3에 axes4를 겹쳐 생성
ax4.set_ylabel("test")
ax4.set_xticks(index + bar_width, minor=True) # 다른 데이터프레임이라 p4의 x 눈금도 표시되는 것을 출력되지 않도록 설정
p4 = ax4.bar(index + bar_width, df['before_val_acc'], bar_width, alpha=alpha, color='r') 
ax4.set_ylim([0.3,0.85])

ax3.legend((p3, p4), ('train', 'test'), fontsize = 12, loc='upper right') # 범례 입력
ax3.set_title("model accuray (train,test)") # Title 입력

plt.show()
```
