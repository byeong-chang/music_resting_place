```python
import pandas as pd
from model_predict import *

song = pd.read_csv("C:/PlayData/music_resting_place/Chat_bot/data/final_music_list.csv")
```

```python
# 토크나이저 로드
def tokenizer_load():
    with open(os.path.join('models', 'tokenizer.pickle'), 'rb') as handle:
        tokenizer = pickle.load(handle)
        return tokenizer
    
# 모델 로드
def predict_model_load():
    predict_model = CatBoostClassifier()
    predict_model.load_model(os.path.join('models', 'catboost.cbm'))
    return predict_model

# 예측
def predict_value(text, model, tokenizer):
    class_names = {4: '슬픔', 3: '상처', 2: '불안', 1: '분노', 0: '기쁨'}
    text_list = preprocess(text)
    list_ = []
    list_.append(text_list)

    # 직접 원-핫 이진 벡터 표현을 얻을 수 있습니다.
    predict_X = tokenizer.texts_to_matrix(list_, mode = 'binary')
    return [class_names[model.predict(predict_X)[0][0]], model.predict_proba(predict_X)]

tokenizer = tokenizer_load()
model = predict_model_load()
```

```python
list_ = []
song['lyrics'].apply(lambda x : list_.append(list(predict_value(x, model, tokenizer)[1][0])))

proba = pd.DataFrame(list_)
cat_proba_lyrics = pd.concat([song, proba], axis = 1)
cat_proba_lyrics.to_csv("./data/cat_proba_lyrics.csv", index = False)
```
